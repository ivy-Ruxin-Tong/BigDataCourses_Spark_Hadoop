```
library = [('Author', 'Topic', 'Pages'), ('Twain', 'Rafting', 601), ('Feynman', 'Physics', 95), ('Hamilton', 'Mythology', 144)]
```

```
library 
```

```
for book in library:
  print(f"topis is {book[1]}")
```

```
 for author, topic, page in library:
    print(f"{author}")
```

```
 for author, topic, page in library:
    print(f"{author} {topic} {page}") # tuple unpacking
```

```
# spacing
for author, topic, page in library:
    print(f"{author:{10}} {topic:{30}} {page:>{10}}") 
```

```
# spacing
for author, topic, page in library:
    print(f"{author:{10}} {topic:{30}} {page:->{10}}")   # 10 caharacters
```

```
# spacing
for author, topic, page in library:
    print(f"{author:{10}} {topic:{30}} {page:.>{10}}") 
```
