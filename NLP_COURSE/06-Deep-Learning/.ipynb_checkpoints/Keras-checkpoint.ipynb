```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
type(iris)
```

```python
print(iris.DESCR)
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
y
```

```python
X
```

```python
# one hot encoding
# class 0 -> [1,0,0]
# class 1 -> [0,1,0]
# class 2 -> [0,0,1]

from keras.utils import to_categorical
```

```python
y = to_categorical(y)
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state =42)
```

```python
y_test # always wanna shuffle the data, otherwise the model is so good at predicting 0, bad at 1 and 2
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
np.array([5,10,15,20]) / 20
```

```python
scalar_object = MinMaxScaler()
```

```python
scaled_X_train = scalar_object.fit_transform(X_train)
```

```python
scaled_X_test = scalar_object.fit_transform(X_test)
```

```python
scaled_X_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8,input_dim = 4, activation = 'relu'))
```

```python
model.add(Dense(8,input_dim = 4, activation = 'relu'))
```

```python
model.add(Dense(3,activation = 'softmax')) # [0.2, 0.3, 0.5]
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train, y_train, epochs = 150, verbose = 1)
#epochs = entire dataset
# verbose, how much output information you want
# verbose = 0, not report anything
# verbose = 1 , progress bar
# verbose = 2, loss, accuracy, details
```

```python
scaled_X_test
```

```python
model.predict(scaled_X_test) # probability 
```

```python
model.predict_classes(scaled_X_test) # predict the label 
```

```python
predictions = model.predict_classes(scaled_X_test)
```

```python
y_test.argmax(axis = 1) # we did the one hot encoding, using this can get it back
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test.argmax(axis = 1), predictions)
```

```python
print(classification_report(y_test.argmax(axis = 1),predictions))
```

```python
accuracy_score(y_test.argmax(axis = 1), predictions)
```
